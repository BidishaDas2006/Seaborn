```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
pivot_table = pd.pivot_table(
    tips,
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean",
    observed=False
)
```

```python


sns.heatmap(
    pivot_table,
    cmap="YlOrRd",
    annot = True
    
)
plt.title("Average Tip by Day and Time")
plt.xlabel("Time")
plt.ylabel("Day")
```
