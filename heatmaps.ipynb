```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Flights = sns.load_dataset("flights")
print(Flights.head())
```

```python
Flights
```

```python
Flights_pivot = Flights.pivot(index="month",columns="year", values="passengers")
Flights_pivot
```

```python
sns.heatmap(
    Flights_pivot,
    cmap="coolwarm",
    annot = True,
    fmt="d"
)

plt.title("passengers Heatmap")
plt.tight_layout()
```
