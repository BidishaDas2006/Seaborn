```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")

```

```python
penguins.head()
```

```python
sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue = "species",
    
)

plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("Flipper Length vs Body Mass (Colored by Species)")

```
