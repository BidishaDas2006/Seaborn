```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
scores = np.random.normal(loc=70, scale=10, size=200)

fig, ax = plt.subplots()

sns.histplot(
    scores,
    ax=ax,
    bins = 30
)
mean_value = np.mean(scores)

ax.axvline(
    x=mean_value,
    color = "red",
    linewidth = "2",
    linestyle = "--",
    
    
)
ax.set_title("Number of students based on scores")
ax.set_xlabel("Scores")
```
