```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue = "sex",
    style = "smoker"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
```
