```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
fig, ax= plt.subplots()

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue = "sex",
    ax=ax
)

ax.set_title("Total Bill For Each Day")
```
